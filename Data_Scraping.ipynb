Here I am using a python library called as AUTOSCRAPER

```python
from autoscraper import AutoScraper
```

```python
url = "https://results.eci.gov.in/PcResultGenJune2024/index.htm"

wanted_list=["Bharatiya Janata Party - BJP","9","0","9"]
```

```python
scraper=AutoScraper()
result =scraper.build(url,wanted_list) 
print(result)
```

```python
scraper.get_result_similar(url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_ap9a':'Party','rule_3z7s':'Won','rule_payx':'Leading','rule_3z7s':'Total'})
scraper.keep_rules(['rule_ap9a','rule_3z7s','rule_payx','rule_3z7s'])
scraper.save('Party Wise Status')
```

```python
scraper.load("Party Wise Status")
result = scraper.get_result_similar("https://results.eci.gov.in/PcResultGenJune2024/partywiseresult-S25.htm", group_by_alias=True)
```

```python
import pandas as pd
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in result.items() ]))
# Converting our DataFrame to CSV file
df.to_csv('Election Dataset/Bihar.csv')
```
